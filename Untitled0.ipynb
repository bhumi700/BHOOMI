```python
import matplotlib.pyplot as plt

countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria']
population = [1439.32, 1380.00, 331.00, 273.52, 220.89, 213.99, 206.14]
plt.figure(figsize=(10, 6))
plt.bar(countries, population, color='skyblue')
plt.xlabel('Countries')
plt.ylabel('Population in Millions')
plt.title('Population of Countries in 2020')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for population in 2020 (in millions)
countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria']
population = [1439.32, 1380.00, 331.00, 273.52, 220.89, 213.99, 206.14]  # population in millions for 2020

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(countries, population, color='skyblue')
plt.xlabel('Countries')
plt.ylabel('Population in Millions')
plt.title('Population of Countries in 2020')
plt.xticks(rotation=45)  # rotate country labels for better readability
plt.tight_layout()

# Show plot
plt.show()
```
